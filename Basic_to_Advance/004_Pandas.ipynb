# Access the Elements in the Data frame

First Way : .loc ===> Series

Second Way : .iloc    ===> Frame

```python
df.loc["Row0"]
```

# Data Series

Data Series means Which has One row or One Column

```python
type(df.loc["Row0"]) ## It is delas with only one row, even though it has multiple column it will consider as datas series because it has only one row
```

# Data Frame

Data frame means it should be more than One row and One Column

```python
df.iloc[:,:]
```

```python
df.iloc[2:,1:]
```

```python
df.iloc[0:2,2:]
```

```python
type(df.iloc[:,:])
```

```python
#Convert DataFrame to Array
df.iloc[:,:].values
```

# CSV

```python
from io import StringIO,BytesIO
```

```python
data = ('col1,col2\n'
         'x,1\n'
         'y,2\n'
         'z,3')
```

```python
type(data)
```

```python
td=pd.read_csv(StringIO(data))
```

```python
td
```

```python
##Read Specifc column 
td=pd.read_csv(StringIO(data),usecols=["col2"])
```

```python
td
```

```python
type(td)
```

```python
## Read the data in the specfic type
td=pd.read_csv(StringIO(data),usecols=["col2"],dtype=int)
```

```python
td
```

```python
td=pd.read_csv(StringIO(data),dtype=object)
```

```python
td
```

```python
td["col2"][1]
```

```python
## Specific the different data type for each column
td=pd.read_csv(StringIO(data),dtype={'col1':object,'col2':float})
```

```python
td
```

```python
td.dtypes
```

```python
## To cahnge the default row index to our custom index from the data
pd.read_csv(StringIO(data),index_col=1)
```
