# Correlation with Heatmap

    A Correlation heatmap uses colored cells, typically in a monochromatic scale , to show 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in feature selection.
    
    Correlation can be find on the floating/integer values not in the categorical values(Object Type)

## Correlation means fidning the inter-relationship between given features

```python
print(df.dtypes)
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Join Plot

A join Plot allows to study the relationship between two numerical variables. The central chart display their correlation. It is usually a scatterplot , a hexbin plot , a 2D histogram or a 2D Density plot

## Bi- Variate Analysis

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind='hex')
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind='reg') ## Draw a regression line
```

## Pair Plot

A "pair plot" is also knowns as  a scatter plot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaboration on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

## Dist Plot

Dist Plot helps us to check the distribution of the column features

```python
sns.displot(df['tip'])
```

```python
sns.displot(df['tip'],kde=False, bins=10)
```

# Categorical Plot

    Seaborn also helps us in doing the analysis on the categorical data points. In this section we will discuss about
        
        Box Plot
        Violin Plot

        bar Plot

## Count Plot

```python
sns.countplot("sex",data=df)
```

```python
sns.countplot("smoker",data=df)
```

```python
sns.countplot("day",data=df)
```

```python
sns.countplot("time",data=df)
```

```python
sns.countplot(y="time",data=df) ## Change the axis direction
```

## Bar Plot

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

## Box Plot

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```
