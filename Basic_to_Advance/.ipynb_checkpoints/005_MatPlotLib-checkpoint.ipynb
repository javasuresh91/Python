# MatplotLib (Simple Visualization Library)

Matplotlib is a plotting library for the Python Programming language and its numerical mathematics extension NumPy. it Porvides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, QT or GTK+.

Some of the major Pros of Matplotlib are : 

    Generally easy to Get started for simple plots
    Support for custom labels and texts
    Great Control of every element in a figure 
    High-Quality output in many formats
    very Customizable in general
   
    

```python
import matplotlib.pyplot as plt

## Basic to show the matplotlib graph we we need to call show(), instead of using it every time, 
## below line will help to show the visualization automatically in Jupyter notebook
%matplotlib inline
```

# Plotting using matploy

```python
##  Plot Scatter ==> Use to scatter the values of x, y axis
plt.scatter(x,y,c="g")
##  Put label for X-asix and Y-axis
plt.xlabel("My X axis")
plt.ylabel("My Y axis")
plt.title("My Plot")
## save the plot as image in any format
plt.savefig("sample_data/test.png")
```

```python
## plt plot ===> We use to draw a line on the plotting points
y = x*x
plt.plot(x,y)
```

# Creating Subplots

```python
plt.subplot(2,2,1) ## First parameter Row count, Second Parameter coulmn count, third parameter position of each graph
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(x,y,'b')

plt.subplot(2,2,4)
plt.plot(x,y,'r*-')
```

# Complex Plot

```python
x = np.arange(0,4 * np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.title("Sin Wave")
```

```python
x = np.arange(0,5 * np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title("Sin")
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g')
plt.title("Cos")
```

# Bar Plot

```python
x = np.arange(0,10)
y = x*x

x2 = np.arange(10,21)
y2 = x2*x2
plt.bar(x,y)
plt.bar(x2,y2,color="g")

plt.show()
```

# Histograms

```python
a = np.array([22,58,2,6,55,48,99,87,23,6,54,89,12,6,96,72,74,23])
plt.hist(a)
```

# Box Plot

```python
a = [np.random.normal(0 , std , 100) for std in range(1,4)]
print(a)
```

```python
plt.boxplot(a,vert=True,patch_artist=True)
```

# Pie Charts

```python
#Data to plot
labels = 'Pyhton' , 'java','c++','Ruby'
size = [215,250,254,658]
colors = ["green","yellow","lightcoral","lightskyblue"]
explode = (0.4,0,0,0) ## This is for part of chart moves

plt.pie(size,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
```
