# 1. Zomato Dataset - Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("zomato.csv",encoding='latin-1')
## encoding is used to read the file in the specific format whenever we get any error while reading the dataset specific to encoding
```

```python
df.head()
```

```python
df.columns
## Just look for the column which is usefult for us or just gather info about it
```

```python
df.info()
## It will give entire information about the columns in dataset
```

```python
df.describe()
## It give the FIVE NUMBER SUMMARY on the Numerical Columns 
```

## In the Data Analysis, We should follow the basic below Things

1. Missing Values
2. Explore about the numerical values
3. Explore about categorical values
4. Finding relationship between features

```python
# 1. Missing Values => Find what are features having null or empty values
df.isnull().sum()
```

```python
## Another way to get Missing values column name 
[feature for feature in df.columns if df[feature].isnull().sum() > 0]
```

```python
## ANother way to get the missing column details
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
## Note : here we cant able to see the missing values because in the dataset we have very very low missing value(only 9)
```

```python
## Reading another suppport file
df_country = pd.read_excel("Country-Code.xlsx")
```

```python
df_country.head()
```

```python
## Combine/Join two data set
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
# To Check the data types
final_df.dtypes
```

```python
final_df.columns
```

```python
## Counting each value in the dataset
final_df.Country.value_counts()
```

```python
## Collecting value name
country_name = final_df.Country.value_counts().index
country_name
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
#Pie Chart Uses the Zomato
plt.pie(x=country_values,labels=country_name)
```

```python
#Pie chart - Top 3 countries uses the Zomato
plt.pie(x=country_values[:3],labels=country_name[:3])
```

```python
## To show Percentage of Data
plt.pie(x=country_values[:3],labels=country_name[:3],autopct='%1.2f%%')
```

###### Observation On the Above Chart : Zomato running the business in the India as MOst and followed by US and UK

```python
# 2. Explore about the relationship between features
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
#Converting to Data frame
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
## Rename the Column of Data frame
rating_info = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
rating_info
```

#### Observation from the above table

1. Whenever rating is between 4.5 to 4.9 ---> Excellent
2. When rating are between 4.0 to 4.5 ----> Very good
3. When rating are between 3.5 to 3.9 ----> Good
4. When rating are between 3.0 to 3.4 ----> Average
5. When rating are between 2.5 to 2.9 ----> Average
6. When rating are between 2.0 to 2.4 ----> Poor

```python
rating_info.head()
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=rating_info)
```

```python
# The charts are looking small, to increase the size
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
```

```python
## Now the chart will be shown big
sns.barplot(x="Aggregate rating",y="Rating Count",data=rating_info)
```

```python
## We color the bar graph with the values in the data iself
sns.barplot(x="Aggregate rating",y="Rating Count",data=rating_info,hue="Rating color",palette=["blue","red","orange","yellow","green","green"])
```

#### Observation from the above chart
1. NOt rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
## COunt plot - Frequency of each rating color in the data set
sns.countplot(x = "Rating color", data=rating_info,palette=["blue","red","orange","yellow","green","green"] )
```

```python
## Find the countries name that have given zero rating
final_df[final_df['Rating color'] == 'White'].groupby(['Country','Rating color']).size().reset_index()
```

#### Observation from above Table:
    1. Indian Customer gave high rating
    2. Western countries gave very low rating

```python
## FInd out which currency is used by which country
final_df.groupby(['Country','Currency']).size().reset_index()
```

```python
## WHich Country have online deliveries option
final_df[final_df['Has Online delivery'] == 'Yes'].groupby(['Country','Has Online delivery']).size().reset_index()
```

```python
## WHich Country doesn't have online deliveries option
final_df[final_df['Has Online delivery'] == 'No'].groupby(['Country','Has Online delivery']).size().reset_index()
```

#### Observation from above tables:
    1. Online Delivery is available in India and UAE
    2. MOst of te country not opt for Online Delivery
