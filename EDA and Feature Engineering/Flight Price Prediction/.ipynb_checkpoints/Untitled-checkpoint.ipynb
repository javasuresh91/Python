#### Observation :
    1. Date of Journey is object(String), but it has three info day,month and year

```python
## Feature Engineering Process
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head() ## New feild added
```

```python
final_df.info()
```

```python
## Converting the Object to Int for date columns(Newly created)
final_df['Date'] =  final_df['Date'].astype(int)
final_df['Month'] =  final_df['Month'].astype(int)
final_df['year'] =  final_df['year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
# On Arrival TIme we observed time and date is combined in the value of it, so need to reomve the date in it
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
## or we can write the same logic in lambda too
## final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
final_df.head(10)
```

```python
## check the null values
final_df.isnull().sum()
```

#### Observation from the null check:
    1. We dont have reasonalbe null values in the feature column
    2. PRoce has null vlaus but it is test data which we appended ealier

```python
## Deriving the new features from Arrtival_time
final_df['Arrival_hour'] = final_df['Arrival_Time'].apply(lambda x: x.split(':')[0])
final_df['Arrival_minutes'] = final_df['Arrival_Time'].apply(lambda x: x.split(':')[1])
##Converting type
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_minutes'] = final_df['Arrival_minutes'].astype(int)
##dropping the arrival_time
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head(10)
```

```python
## Deriving the new features from Departure_time
final_df['Departure_hour'] = final_df['Dep_Time'].apply(lambda x : x.split(':')[0])
final_df['Departure_minutes'] = final_df['Dep_Time'].apply(lambda x: x.split(':')[0])
final_df['Departure_hour'] = final_df['Departure_hour'].astype(int)
final_df['Departure_minutes'] = final_df['Departure_minutes'].astype(int)
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df.head(10)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] =  final_df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, 'nan':2, '4 stops':4})
```

```python
final_df.drop('Route',axis=1,inplace=True)
final_df.head()
```

```python
## Duration Column does not have consistent values, so we can drop off
final_df[final_df['Duration'] == '5m']
```

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_hour'] = final_df['Duration'].apply(lambda x : x.split(' ')[0].split('h')[0])
final_df['Duration_hour'] = final_df['Duration_hour'].astype(int)
final_df.drop('Duration',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Airline'].unique()
```

```python
#ANother techinque for alternative for dummies or giving the numerical values manually using map
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Source'] = labelencoder.fit_transform(final_df['Source'])
final_df['Destination'] = labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = labelencoder.fit_transform(final_df['Additional_Info'])
```
