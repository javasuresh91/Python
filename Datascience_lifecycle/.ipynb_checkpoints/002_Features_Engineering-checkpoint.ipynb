## Missing Value

```python
## Find the Null based features for the categorical features

category_feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1 and dataset[feature].dtype == 'O']
print(category_feature_nan)
## Just analysis for our understanding
a = [s for s in dataset['MasVnrType'].isnull() if s == False]
b = [s for s in dataset['MasVnrType'].isnull() if s == True]
print(len(a) , len(b))
print(dataset['MasVnrType'].isnull().mean())
for fe in category_feature_nan:
    print("{} : missing {} % ".format(fe,np.round(dataset[fe].isnull().mean(),4)))
```

```python
## Replacing the missing value with a new label(value)
def replace_nan(dataset , feature_nan):
    data = dataset.copy()
    data[feature_nan] = data[feature_nan].fillna('Missing')
    return data

dataset = replace_nan(dataset,category_feature_nan)
dataset[category_feature_nan].isnull().sum() ## Converted value
```

```python
dataset
```

```python
numerical_feature_nan = [feat for feat in dataset.columns if dataset[feat].isnull().sum() > 1 and  dataset[feat].dtypes !='O' ]
print("Count of Numerical features Nan : {}".format(len(numerical_feature_nan)))
```

```python
for nv in numerical_feature_nan:
    print("{} : {} % missing".format(nv,np.around(dataset[nv].isnull().mean(),4)))
```

```python
## Replacing Nan in the Numerical features
for fe in numerical_feature_nan:
    median_value = dataset[fe].median()
    dataset[fe].fillna(median_value,inplace=True)    
    ## create a new feature to capture some additional new info
    dataset[fe+"_nan"] = np.where(dataset[fe].isnull(),1,0)
for nv in numerical_feature_nan:
    print("{} : {} % missing".format(nv,np.around(dataset[nv].isnull().mean(),4)))
```

## Temporal Variable

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt','YrSold']]
```

```python
for fe in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[fe] = dataset['YrSold']-dataset[fe]
```

```python
dataset
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt','YrSold']]
```

## Numerical Variable

```python
dataset
```

```python
num_features = ["LotFrontage","LotArea","1stFlrSF","GrLivArea","SalePrice"]

for f in num_features:
    dataset[f] = np.log(dataset[f])
```

```python
dataset.head()
```

## Handling Rare Categorical Features

We will remove categorical variable that are present less than 1% of the observation

```python
categorical_features = [fe for fe in dataset.columns if dataset[fe].dtype == 'O']
```

```python
categorical_features
```

```python
for feat in categorical_features:
    temp = dataset.groupby(feat)['SalePrice'].count()/len(dataset)
    temp_df = temp[temp>0.01].index
    dataset[feat] = np.where(dataset[feat].isin(temp_df),dataset[feat],'Rare_var')
```

```python
dataset
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

## Feature Scaling

```python
feature_scaling = [fe for fe in dataset.columns if fe not in ['Id','SalePrice']]
print(len(feature_scaling))
print(feature_scaling)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset[feature_scaling])
```

```python
dataset
```

```python
scaler.transform(dataset[feature_scaling])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scaling]), columns=feature_scaling)],
                    axis=1)
```

```python
data.head()
```

```python
data.to_csv('X_train.csv',index=False)
```
